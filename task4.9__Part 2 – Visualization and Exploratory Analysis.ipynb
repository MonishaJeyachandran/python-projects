```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Load  dataframe
file_path = r"C:\Users\richa\Documents\01-2026 Instacart Basket Analysis\02 Data\Prepared Data\ords_prods_merged_customers.pkl"
ords_prods_merge = pd.read_pickle(file_path)
```

```python
#column names
ords_prods_merge.columns
```

```python
#Histogram
# Histogram
hist = ords_prods_merge['order_hour_of_day'].plot.hist(bins=24, color='skyblue')
plt.title('Orders by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
```

```python
# Export
plt.savefig(os.path.join(r"C:\Users\richa\Documents\01-2026 Instacart Basket Analysis\04 Analysis\Visualizations", 'hist_order_hour.png'))
plt.show()
```

```python
#Most orders occur between 9:30 and 16:00, suggesting busy shopping times. 
#Marketing could target promotions during these hours.
```

```python
# 4 Bar chart
bar = ords_prods_merge['loyalty_flag'].value_counts().plot.bar(color=['purple','orange','green'])
plt.title('Customer Loyalty Distribution')
plt.xlabel('Loyalty Level')
plt.ylabel('Number of Customers')
```

```python
#The bar chart shows regular customer>loyal customer >new customer. This helps marketing target campaigns based on loyalty.
```

```python
#5 Line chart-hour of the day
df_sample = ords_prods_merge.sample(n=2_000_000, random_state=4)

plt.figure(figsize=(12,6))
sns.lineplot(x='order_hour_of_day', y='prices', data=df_sample)
plt.title('Average Expenditure by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Price ($)')
plt.show()
```

```python
#6 age and no of dependents
plt.figure(figsize=(12,6))
sns.lineplot(x='age', y='n_dependents', data=ords_prods_merge)
plt.title('Number of Dependents by Customer Age')
plt.xlabel('Age')
plt.ylabel('Number of Dependents')
plt.show()
```

```python
 #7 scatter plot : age and income
plt.figure(figsize=(12,6))
sns.scatterplot(x='age', y='income', data=ords_prods_merge, alpha=0.5)
plt.title('Income vs Customer Age')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.show()
```

```python
#export
plt.figure(figsize=(10,5))
sns.histplot(ords_prods_merge['order_hour_of_day'], bins=24, color='skyblue')
plt.title('Distribution of Orders by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.savefig(os.path.join(file_path, '04 Analysis', 'order_hour_histogram.png'))
```
