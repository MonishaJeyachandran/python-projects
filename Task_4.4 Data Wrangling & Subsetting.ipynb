```python
#4.4 Data Wrangling & Subsetting
#Import Libraries
import pandas as pd
import numpy as np
import os

# 1.Set the path to your Original Data folder (central location)
data_path=r'C:\Users\richa\Documents\01-2026 Instacart Basket Analysis\02 Data\Original Data'

# Load CSV files using os.path.join into pandas dataframes
df_ords = pd.read_csv(os.path.join(data_path, 'orders.csv'), index_col=False)
df_prods = pd.read_csv(os.path.join(data_path, 'products.csv'),index_col=False)
df_dep = pd.read_csv(os.path.join(data_path, 'departments.csv'), index_col=False)
```

```python
#2.change IDs to strings: Convert order_id and user_id to string
df_ords['order_id'] = df_ords['order_id'].astype('str')
df_ords['user_id'] = df_ords['user_id'].astype('str')
```

```python
# verfication
df_ords['order_id'].dtype  
df_ords['user_id'].dtype
```

```python
# 3.Drop the 'eval_set' column 
df_ords=df_ords.drop (columns=['eval_set'])
```

```python
# 4. Busiest hour for placing orders
df_ords['order_hour_of_day'].value_counts()
```

```python
# 5. meaning behind a value of 4 in the "department_id"
#from excercise
# Transpose the departments dataframe
df_dep_t = df_dep.T

# Reset the index to make the first row a proper row
df_dep_t = df_dep_t.reset_index()

# Take the first row to use as header
new_header = df_dep_t.iloc[0]

# Create a new dataframe without the first row
df_dep_t_new = df_dep_t[1:]

# Assign the new header
df_dep_t_new.columns = new_header

# Check the cleaned dataframe
df_dep_t_new.head()

# Convert the cleaned dataframe to a dictionary using row index as key
data_dict = df_dep_t_new.set_index('department_id')['department'].to_dict()

# Check the dictionary
print(data_dict)
```
